```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import mplcyberpunk
```

```python
ativos = ["^BVSP", "^GSPC", "TAEE11.SA", "BBAS3.SA", "ITSA4.SA", "PETR4.SA", "WEGE3.SA", "HGLG11.SA", "KNCR11.SA", "KNRI11.SA", "XPML11.SA", "AAPL", "SPG", "O", "BTC=F"]
```

```python
dados_mercado = yf.download(ativos, period = "6mo")
dados_mercado = dados_mercado["Close"]
dados_mercado
```

```python
dados_mercado = dados_mercado.dropna()

dados_mercado
```

```python
dados_mercado.columns = ["Apple", "BBAS3", "Bitcoin", "HGLG11", "ITSA4", "KNCR11", "KNRI11", "O Income Realty", "PETR4", "Simon Group", "TAEE11", "WEGE3", "XPML11", "Ibovespa", "S&P 500" ]

dados_mercado
```

```python
plt.style.use("cyberpunk")
```

```python
plt.plot(dados_mercado["O Income Realty"])
plt.title("Ações O 6M")

plt.savefig("O.png")
```

```python
plt.plot(dados_mercado["Apple"])
plt.title("Ações Apple 6M")

plt.savefig("Apple.png")
```
